# Organize Netflix movie actor rating data

## 1. Analysis Objective

The purpose of this data analysis is to organize actor rating data for films and TV works across different genres, such as comedy, action, and sci-fi. By calculating the average IMDb rating of the works each actor has appeared in within each genre, this analysis aims to identify actors associated with high-rated works in each genre.

The purpose of this practical project is to practice data cleaning and organization, so as to obtain a dataset that can be used for further analysis.

## 2. Introduction

The original dataset records all Netflix movies and TV shows available in the United States as of July 2022. The dataset contains two data tables: `titles.csv` and `credits.csv`.

`titles.csv` contains information related to movies and TV shows, including the title ID, title, type, description, genres, IMDb rating, and other attributes. IMDb is an international online platform for film and TV ratings. `credits.csv` contains information on more than 70,000 directors and actors who appeared in Netflix titles, including their names, title IDs, character names, and roles in the production, such as director or actor.

The meanings of each column in `titles.csv` are as follows:

- `id`: The ID of the film or TV title.  
- `title`: The title of the film or TV show.  
- `type`: The type of title, either a TV show or a movie.  
- `description`: A brief description.  
- `release_year`: The release year.  
- `age_certification`: The age rating.  
- `runtime`: The duration of each TV episode or movie.  
- `genres`: A list of genres.  
- `production_countries`: A list of production countries.  
- `seasons`: The number of seasons, if the title is a TV show.  
- `imdb_id`: The IMDb ID.  
- `imdb_score`: The IMDb rating.  
- `imdb_votes`: The number of IMDb votes.  
- `tmdb_popularity`: The TMDB popularity score.  
- `tmdb_score`: The TMDB rating.  
  
The meanings of each column in `credits.csv` are as follows:  
  
- `person_id`: The ID of the cast or crew member.  
- `id`: The ID of the film or TV title they participated in.  
- `name`: The name of the person.  
- `character`: The name of the character.  
- `role`: The role type, either actor or director.

## Data evaluation and cleaning

### Back up original data

```python
cleaned_titles = original_titles.copy()
cleaned_credits = original_credits.copy()
```

### Evaluation and cleaning of `titles.csv`

#### Data tidiness

Each column represents a variable and each row represents an observation. However, this project aims to organize the average IMDb scores of works across different genres and actors. Therefore, the lists in the `genres` column need to be split into separate rows. However, since there is no additional requirement related to countries, the lists in the `production_countries` column do not need to be split.

Convert the `genres` column from string format into actual Python lists。

```python
cleaned_titles['genres'] = cleaned_titles['genres'].apply(lambda s: eval(s)) 
```

After converting the column into lists, we can use the DataFrame explode() method to split the list values in that column into separate rows, which makes the data easier to clean and arrange later.

```python
cleaned_titles = cleaned_titles.explode("genres")
cleaned_titles
```

##### Missing data

Remove rows with missing `genres` or `imdb_score` values, and then check the remaining number of missing values in these columns.

```python
cleaned_titles.dropna(subset = "genres", inplace = True)
cleaned_titles.dropna(subset = "imdb_score", inplace = True)
print(cleaned_titles['genres'].isna().sum())
print(cleaned_titles['imdb_score'].isna().sum())
```

## Merge DataFrames

Both DataFrames have the column `id`, `cleaned_titles` and `cleaned_credits` should be merged based on the column `id`.

```python
merged_df = pd.merge(cleaned_titles, cleaned_credits, on = "id", how = "inner")
merged_df
```

## Performing groupby operations on the DataFrame.

Create a pivot table to calculate the average IMDb score for each actor within each genre.

```python
initial_results = pd.pivot_table(merged_df, index = ["genres","person_id"], values = 'imdb_score', aggfunc = np.mean)
initial_results
```

Reset the index to convert `genres` and `person_id` back into regular columns, making the result easier to sort, filter, and analyze.

```python
results_reset_index = initial_results.reset_index()
results_reset_index
```

For each genre, find the highest average IMDb score among all actors.

```python
genres_max_scores = results_reset_index.groupby("genres")["imdb_score"].max()
genres_max_scores
```

Merge the maximum scores back to the actor-level results to find which actors achieved the highest average IMDb score in each genre.

```python
genres_max_scores_person_id = pd.merge(results_reset_index, genres_max_scores, on = ["genres", "imdb_score"])
genres_max_scores_person_id
```

Create a reference table that maps each `person_id` to the corresponding actor `name`, so that actor names can be merged back into the final result. Duplicated rows were removed to avoid repeated matches during the merge. The `person_id` column was also checked to ensure that each actor ID appears only once in this reference table.

```python
person_id_names = cleaned_credits[["person_id", "name"]].drop_duplicates(subset=["person_id"])
person_id_names
```

Merge the actor names into the maximum-score result table, sort the results by `genres`, reset the index, and remove the old index column.

```python
genres_max_score_names = pd.merge(genres_max_scores_person_id, person_id_names, on = 'person_id').sort_values('genres').reset_index().drop('index', axis = 1)
genres_max_score_names
```
