# Cleaning and Assessing E-commerce Transaction Data

## 1. Analysis Objective

The purpose of this data analysis project is to identify best-selling products based on sales transaction data, so that more effective marketing strategies can be developed to increase revenue.

The purpose of this practice project is to assess the cleanliness and tidiness of the dataset, clean the data based on the assessment results, and prepare the dataset for further analysis.

## 2. Introduction

The original dataset records all transactions of a UK-based online retail company from December 1, 2010 to December 9, 2011. It covers the company’s business data across different countries and regions. The company mainly sells gifts for various occasions, including but not limited to birthday gifts, wedding souvenirs, Christmas gifts, and more. Its customer groups mainly include wholesalers and individual consumers, with wholesalers accounting for a relatively large proportion.

The meaning of each column is as follows:

- `InvoiceNo`: Invoice number. A 6-digit number that uniquely identifies each transaction. If this code starts with the letter "C", it indicates that the transaction was cancelled.
- `StockCode`: Product code. A 5-digit code that uniquely identifies each product.
- `Description`: Product name.
- `Quantity`: The quantity of the product in the transaction.
- `InvoiceDate`: Invoice date and time. The date and time when the transaction occurred.
- `UnitPrice`: Unit price. The price of the product in British pounds (£).
- `CustomerID`: Customer number. A 5-digit number that uniquely identifies each customer.
- `Country`: Country name. The country where the customer is located.

Import the library needed for data analysis, then use pandas `read_csv()` function to load the raw data file `e_commerce.csv` as a DataFrame named `original_data`.

```python
import pandas as pd
original_data = pd.read_csv("./e_commerce.csv")
original_data
```

## 4. Data Assessment

In this section, I will assess the `original_data` DataFrame created in the previous section.

The assessment focuses on data structure and data quality. Structural issues are checked based on tidy data principles: each column should represent one variable, each row should represent one observation, and each cell should contain one value. Data quality issues include missing values, duplicated records, invalid values, and other problems that may affect the analysis.

### 4.1 Data Preview and tidiness assessment

```python
original_data.sample(10)
```

From a random sample of 10 rows, each column represents a variable, each row represents an observation, and each cell contains a single value. Therefore, the dataset is tidy.

##### 4.2.1.2 `CustomerID` assessment

```python
original_data[original_data["CustomerID"].isnull()]
```

There is a many-to-many relationship between customers and products: one customer can purchase multiple products, and one product can be purchased by multiple customers. ##### 4.2.4.2 `UnitPrice` Assessment

## 5. Data cleaning

### 5.1 Items to Clean

- Convert the data types of `InvoiceDate` and `CustomerID`
- Remove rows with missing `Description`
- Remove duplicated rows
- Standardize the country names for `United Kingdom` and `United States`
- Remove rows with negative `Quantity`
- Remove rows with negative `UnitPrice`

### 5.2 Back Up Original Data

```python
cleaned_data = original_data.copy()
```

Back up the original dataset to avoid losing the raw data during cleaning.

### 5.3 Data type conversion

#### 5.3.1 Conversion of `InvoiceDate`

```python
cleaned_data["InvoiceDate"] = pd.to_datetime(cleaned_data["InvoiceDate"])
```

Convert `InvoiceDate` from a string data type to a datetime data type.

#### 5.3.2 Conversion of `CustomerID`

```python
cleaned_data["CustomerID"] = cleaned_data["CustomerID"].astype("Int64").astype(str)
```

Convert `CustomerID`from a float data type to a string data type.

Check the format of converted `CustomerID`.

```python
cleaned_data["CustomerID"]
```

The converted `CustomerID` still included ".0", which is not reasonable in case of `CustomerID` and should be removed.

### 5.4 Remove rows with missing `Description`

```python
cleaned_data.dropna(subset = ["Description"], inplace = True)
```

### 5.5 Remove duplicated rows

```python
cleaned_data.drop_duplicates(inplace = True)
```

### 5.6 Standardize the country names for United Kingdom and United States

```python
cleaned_data.replace(['UK', 'U.K.'], 'United Kingdom', inplace = True)
cleaned_data.replace('USA', 'United States', inplace = True)
```

Check the number of "UK", "U.K" and "USA" after replacement.

```python
print(len(cleaned_data[cleaned_data["Country"] == "UK"]))
print(len(cleaned_data[cleaned_data["Country"] == "U.K."]))
print(len(cleaned_data[cleaned_data["Country"] == "USA"]))
```

After standardization, there are no remaining records with country names written as `UK`, `U.K`, or `USA`.

#### 5.6 Remove rows with negative `Quantity`

```python
cleaned_data = cleaned_data[cleaned_data["Quantity"] >= 0]
```

Check the number of negative `Quantity` after replacement.

```python
len(cleaned_data[cleaned_data["Quantity"] < 0])
```

After standardization, there are no remaining records with negative `Quantity`.

#### 5.7 Remove rows with negative `UnitPrice`

```python
cleaned_data = cleaned_data[cleaned_data["UnitPrice"] >= 0]
```

Check the number of negative `UnitPrice` after replacement.

```python
len(cleaned_data[cleaned_data["UnitPrice"] < 0])
```

After standardization, there are no remaining records with negative `UnitPrice`.

Save the cleaned dataset as `cleaned_data.csv` in the same folder.
